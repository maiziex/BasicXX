```matlab
Share_IBD=5/8;
Share_nonIBD=0.5;
```

```matlab
 result=[];
 simdata=rand(3000,1000)>=0.9;
 for i=1:3:3000
     ind1=simdata(i,:)+simdata(i+1,:);
     ind2=simdata(i+1,:)+simdata(i+2,:);
          [row_1,col_1]=find(ind1==0&ind2~=0);
     [row_2,col_2]=find(ind1~=0&ind2==0);
     [row_3,col_3]=find(ind1==0&ind2==0);
     num_ns=size(col_1,2)+size(col_2,2)+size(col_3,2);
     sumtemp=(1000-num_ns)*Share_IBD+num_ns*Share_nonIBD;
     result=[result;sumtemp];
end
%simulation for IBD region
hist(result,10);
xlabel('Likelihood test')
```
